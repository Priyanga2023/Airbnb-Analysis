```python
data = []
for i in col.find({},{'_id':1,'listing_url':1,'name':1,'accommodates':1,'property_type':1,'room_type':1,'bed_type':1,'bedrooms': 1,'beds':1,
              'bathrooms':1,'price':1,'extra_people':1,'guests_included':1,'minimum_nights':1,'maximum_nights':1,'security_deposit':1,
              'cancellation_policy':1,'cleaning_fee':1,'number_of_reviews':1,'review_scores.review_scores_rating':1,'images.picture_url':1}):
    data.append(i)
df = pd.DataFrame(data)
df['images'] = df['images'].apply(lambda x: x['picture_url'])
df['review_scores'] = df['review_scores'].apply(lambda x: x.get('review_scores_rating',0))
df.head()
```

```python
df.isnull().sum()
```

```python
df['bedrooms'].fillna(0,inplace=True)
df['beds'].fillna(0,inplace=True)
df['bathrooms'].fillna(0,inplace=True)
df['cleaning_fee'].fillna('Not Specified',inplace=True)
df['security_deposit'].fillna('Not Specified',inplace=True)

df.isnull().sum()     
```

```python
df.dtypes
```

```python
df['bedrooms'] = df['bedrooms'].astype(int)
df['beds'] = df['beds'].astype(int)
df['bathrooms'] = df['bathrooms'].astype(str).astype(float)
df['minimum_nights'] = df['minimum_nights'].astype(int)
df['maximum_nights'] = df['maximum_nights'].astype(int)
df['price'] = df['price'].astype(str).astype(float).astype(int)
df['security_deposit'] = df['security_deposit'].apply(lambda x: int(float(str(x))) if x != 'Not Specified' else 'Not Specified')
df['cleaning_fee'] = df['cleaning_fee'].apply(lambda x: int(float(str(x))) if x != 'Not Specified' else 'Not Specified')
df['extra_people'] = df['extra_people'].astype(str).astype(float).astype(int)
df['guests_included'] = df['guests_included'].astype(str).astype(int)
```

```python
# Address
```

```python
address = []
for i in col.find({},{'_id':1,'address':1}):
    address.append(i)

df_address = pd.DataFrame(address)
address_keys = list(df_address.iloc[0,1].keys())

for key in address_keys:
    if key == 'location':
        df_address['location_type'] = df_address['address'].apply(lambda x: x.get('location', {}).get('type', 'Not Specified'))
        df_address['longitude'] = df_address['address'].apply(lambda x: x.get('location', {}).get('coordinates', [])[0])
        df_address['latitude'] = df_address['address'].apply(lambda x: x.get('location', {}).get('coordinates', [])[1])
        df_address['is_location_exact'] = df_address['address'].apply(lambda x: x.get('location', {}).get('is_location_exact', 'Not Specified'))
    else:
        df_address[key] = df_address['address'].apply(lambda x: x.get(key, 'Not Specified') if x and key in x and x[key] != '' else 'Not Specified')
        
df_address.drop(columns=['address'], inplace=True)
df_address.head() 
```

```python
df_address.isnull().sum()
```

```python
df_address.dtypes
```

```python
# Amenities
```

```python
amenities = []
for i in col.find({},{'_id': 1,'amenities': 1}):
    amenities.append(i)

df_amenities = pd.DataFrame(amenities)
df_amenities['amenities_count'] = df_amenities['amenities'].apply(lambda x: len(str(x).split(',')))
df_amenities.head()
```

```python
# Availability
```

```python
availability = []
for i in col.find({},{'_id':1,'availability':1}):
    availability.append(i)

df_availability = pd.DataFrame(availability)

df_availability['availability_30'] = df_availability['availability'].apply(lambda x: x.get('availability_30', 0))
df_availability['availability_60'] = df_availability['availability'].apply(lambda x: x.get('availability_60', 0))
df_availability['availability_90'] = df_availability['availability'].apply(lambda x: x.get('availability_90', 0))
df_availability['availability_365'] = df_availability['availability'].apply(lambda x: x.get('availability_365', 0))

df_availability.drop(columns=['availability'], inplace=True)
df_availability.head()
```

```python
df_availability.isnull().sum()
```

```python
df_availability.dtypes
```

```python
# Host
```

```python
host = []
for i in col.find({},{'_id': 1,'host': 1}):
    host.append(i)

df_host = pd.DataFrame(host)
host_keys = list(df_host.iloc[0, 1].keys())
host_keys.remove('host_about')

for key in host_keys:
    if key == 'host_response_time':
        df_host['host_response_time'] = df_host['host'].apply(lambda x: x.get('host_response_time', 'Not Specified'))
    else:
        df_host[key] = df_host['host'].apply(lambda x: x.get(key, 'Not Specified') if x and key in x and x[key] != '' else 'Not Specified')

df_host.drop(columns=['host'], inplace=True)
df_host.head()
```
